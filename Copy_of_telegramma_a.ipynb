```python
import re

def restore_whitespace_and_punctuation(text):

    matches = re.findall(r'\s|[.,!?]', text)


    replacements = {
        ' ': ' ',  # Пробел
        '\n': '\n',  # Новая строка
        '.': '.',  # Точка
        ',': ',',  # Запятая
        '!': '!',  # Восклицательный знак
        '?': '?'  # Вопросительный знак
    }


    restored_text = []
    index = 0
    for i, char in enumerate(text):
        if char in replacements:
            restored_text.append(replacements[char])
        else:
            restored_text.append(char)

    return ''.join(restored_text)


with open('/telegramma_text.txt', 'r', encoding='utf-8') as file:
    raw_text = file.read()
    restored_text = restore_whitespace_and_punctuation(raw_text)
    print(restored_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import re


def load_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text_cleaned = file.read()
        print(f"Текст из файла:\n{text_cleaned}\n")
    return text_cleaned

file_path = '/content/step3_result.txt'
text_cleaned = load_text(file_path)


stop_words = set(stopwords.words('russian'))
stopwords_re = r'\b(?:{})\b'.format('|'.join(stop_words))
cleaned_text = re.sub(stopwords_re, '', text_cleaned, flags=re.IGNORECASE)


vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([cleaned_text])


feature_names = vectorizer.get_feature_names_out()
densified_tfidf = tfidf_matrix.todense().tolist()[0]


keywords = []
for col in range(len(feature_names)):
    keywords.append((feature_names[col], densified_tfidf[col]))


sorted_keywords = sorted(keywords, key=lambda x: x[1], reverse=True)


top_n = 10
print(f"Топ-{top_n} ключевых слов:")
for i in range(top_n):
    print(sorted_keywords[i][0], f"(TF-IDF: {sorted_keywords[i][1]:.4f})")
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary
import nltk


file_path = '/content/step3_result.txt'  # Укажите правильный путь к вашему файлу
with open(file_path, 'r', encoding='utf-8') as file:
    text_cleaned = file.read()


sentences = text_cleaned.split('.')


stop_words = nltk.corpus.stopwords.words('russian')
def preprocess(sentence):
    result = []
    for token in simple_preprocess(sentence):
        if token not in stop_words and len(token) > 3:
            result.append(token)
    return result

processed_sentences = [preprocess(sent) for sent in sentences]

dictionary = Dictionary(processed_sentences)
corpus = [dictionary.doc2bow(text) for text in processed_sentences]

lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=10,  # Количество тем
    random_state=100,
    update_every=1,
    chunksize=100,
    passes=20,
    alpha=1,
    per_word_topics=True
)

topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)
```

```python
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny-sentiment-balanced")
model = AutoModelForSequenceClassification.from_pretrained("cointegrated/rubert-tiny-sentiment-balanced")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

def estimate_sentiment(sentences):
    inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors="pt").to(device)

    outputs = model(**inputs)
    logits = outputs.logits
    probabilities = torch.softmax(logits, dim=1).detach().cpu().numpy()

    sentiment_scores = []
    for prob in probabilities:
        sentiment_score = {
            "negative": round(prob[0].item(), 4),
            "neutral": round(prob[1].item(), 4),
            "positive": round(prob[2].item(), 4)
        }
        sentiment_scores.append(sentiment_score)

    return sentiment_scores

sentences = [
    "Это отличное предложение!",
    "Это нейтральное предложение.",
    "Это плохое предложение."
]

sentiment_scores = estimate_sentiment(sentences)

average_negative = sum(score["negative"] for score in sentiment_scores) / len(sentiment_scores)
average_neutral = sum(score["neutral"] for score in sentiment_scores) / len(sentiment_scores)
average_positive = sum(score["positive"] for score in sentiment_scores) / len(sentiment_scores)

for index, (sentence, score) in enumerate(zip(sentences, sentiment_scores)):
    print(f"Предложение {index+1}: {sentence}")
    print(f"Оценка тональности: {score}\n")

print("Средняя оценка тональности:")
print(f"Отрицательная: {average_negative:.4f}")
print(f"Нейтральная: {average_neutral:.4f}")
print(f"Положительная: {average_positive:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict


with open('/content/new_file.txt', 'r') as file:
    text = file.read().strip()

names = ['настя', 'катерина', 'тимофеев', 'тихон', 'манюшка']

interactions = defaultdict(int)

for name in names:
    for other_name in names:
        if name != other_name:
            interactions[(name, other_name)] += text.count(name + ' ' + other_name)

G = nx.Graph()

for (u, v), weight in interactions.items():
    G.add_edge(u, v, weight=weight)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)
nx.draw_networkx_labels(G, pos, font_size=12)
nx.draw_networkx_edges(G, pos, edge_color='gray', width=[weight*2 for u,v,weight in G.edges(data='weight')])
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels,font_size=10)

plt.axis('off')
plt.show()
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary
import nltk

def load_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text_cleaned = file.read()
        print(f"Текст из файла:\n{text_cleaned}\n")
    return text_cleaned

file_path = '//content/saved_cleaned_text.txt'
texts = load_text(file_path)

sentences = texts.split('.')

stop_words = nltk.corpus.stopwords.words('russian')
def preprocess(sentence):
    result = []
    for token in simple_preprocess(sentence):
        if token not in stop_words and len(token) > 3:
            result.append(token)
    return result

processed_sentences = [preprocess(sent) for sent in sentences]

dictionary = Dictionary(processed_sentences)
corpus = [dictionary.doc2bow(text) for text in processed_sentences]

lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=10,
    random_state=100,
    update_every=1,
    chunksize=100,
    passes=10,
    alpha='auto',
    per_word_topics=True
)

topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)
```
